```python
import math
import cv2
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
def showImage(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    figure(figsize=(5, 5))
    plt.imshow(image)
    plt.show()    
```

```python
path = r"/home/dan/8sem/DIP/Lab2/mandarins1.png"
minD = 90
minR = 55
maxR = 95
```

```python
image = cv2.imread(path)
showImage(image)
```

```python
res = image.copy()
gray = cv2.cvtColor(res, cv2.COLOR_BGR2GRAY)
circles = cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT, 2, minDist = minD, minRadius = minR, maxRadius = maxR)
circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    cv2.circle(res,(i[0],i[1]),i[2],(0,255,0),2)
    cv2.circle(res,(i[0],i[1]),2,(0,0,255),3)
showImage(res)
```
